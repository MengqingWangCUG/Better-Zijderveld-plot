```python
import sys
import os
import importlib
current_dir = os.getcwd()
if current_dir not in sys.path:
    sys.path.insert(0, current_dir)  
modules_to_remove = [key for key in sys.modules.keys() if key.startswith('test')]
for module in modules_to_remove:
    del sys.modules[module]
import importlib.util
spec = importlib.util.spec_from_file_location("test", os.path.join(current_dir, "test.py"))
test = importlib.util.module_from_spec(spec)
spec.loader.exec_module(test)

import numpy as np
np.random.seed(42)
test.run_all_tests()
```
